```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://aiagentsdirectory.com'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url + "/landscape", headers)
if(response.status_code != 200):
    print(url + "/landscape" + " : failed")
    
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
with open("./ai_agents_directory.txt", "w") as text_file:
    text_file.write("ai agents directory\n")

for atag in soup.find_all(class_='agent-link', href=True):
    link = atag.get('href')
    try:
        data = scrape_product_page(link)
        line = write_line(data)
        with open("./ai_agents_directory.txt", "a") as text_file:
            text_file.write(line)
        time.sleep(2)
        print(link)
    except:
        print(link + " : failed")
```

```python
# Name
# Website
# Category
# Pricing
# Description
# Use Cases

def scrape_product_page(endpoint: str):
    data = {}
    response = requests.get(url + endpoint, headers)
    if(response.status_code != 200):
        print(url + "/landscape" + " : failed")
        
    soup = BeautifulSoup(response.content, 'html.parser')
    data['name'] = str(soup.find('title')).split(' - ')[0][7:]
    data['website'] = soup.find_all(class_='hover:bg-green-100', href = True)[0].get('href')
    
    soup.find_all('p', class_='text-gray-600')
    category_tag = list(filter(lambda x: 'Category:' in str(x), soup.find_all('p', class_='text-gray-600')))
    data['category'] = str(category_tag[0]).split('>')[2].split('<')[0]
    
    soup.find_all('p', class_='text-gray-600')
    category_tag = list(filter(lambda x: 'Pricing Model' in str(x), soup.find_all('p', class_='text-gray-600')))
    data['pricing'] = str(category_tag[0]).split('>')[2].split('<')[0]
    
    data['description'] = str(soup.find_all('p', class_='text-gray-600 leading-relaxed')[0]).split('>')[1][:-3].replace("\n", " ").replace("\r", " ").strip()
    
    use_case_str = ""
    for bullet in soup.find_all('div', class_='rounded-lg border bg-card text-card-foreground shadow-sm')[2].find_all('span'):
        use_case_str += str(bullet).split('>')[1].split('<')[0] + ' / '
    data['use_case'] = use_case_str.replace("\n", " ").replace("\r", " ").strip()

    return data
```

```python
def write_line(data_dict: dict()):
    line = ""
    for i in data_dict:
        line += data_dict[i] + " | "
    line += "\n"
    return line
```
